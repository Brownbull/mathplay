```
# Introduction https://docs.sympy.org/latest/modules/geometry/index.html
x1 = Point(0, 0)
x2 = Point(1, 1)
x3 = Point(2, 2)
zp = Point(1, 0)
print('Collinearity:', Point.is_collinear(x1, x2, x3))
print('Collinearity:', Point.is_collinear(x1, x2, zp))
t = Triangle(zp, x2, x1)
print('Triangle Area: ', t.area)
print('Triangle Medians: ', t.medians[x1])
m = t.medians
intersection(m[x1], m[x2], m[zp])
```
