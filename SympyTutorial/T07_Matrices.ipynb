```
# MATRICES https://docs.sympy.org/latest/tutorial/matrices.html
Matrix([[1, -1], [3, 4], [0, 2]])
```

```
# Vectors
Matrix([1,2,3])
```

```
# Mulitplication
M = Matrix([[1,2,3], [3,2,1]])
N = Matrix([0,1,1])
M*N
```

```
# shape
M = Matrix([[1,2,3],[-2,0,4]])
print "Shape: \t", M.shape
M
```

```
# Rows and columns
M.row(0)
```

```
M.col(-1)
```

```
# Deleting columns rows
M.col_del(0)
M
```

```
#Inserting columns rows
M.row_del(1)
M
```

```
M = M.row_insert(1, Matrix([[0,4]]))
M
```

```
M = M.col_insert(0, Matrix([1, -2]))
M
```

```
# Basics operations
M = Matrix([[1,3],[-2,3]])
N = Matrix([[0,3],[0,7]])
print(M + N)
print(M * N)
print(M ** 2)
```

```
# Inverse of a Matrix
M**-1
```
