```
# Manipulating symbols functions
from sympy import symbols, expand, factor
x, y = symbols('x y')
f = x + 2*y
print('f 1: ', f)
f = f + 1
print('f 2: ', f)
f = f - x
print('f 3: ', f)
f = (f + x - 1)*x
print('f 4: ', f)
expand_f = expand(f)
print('f 4 expand: ', expand_f)
factor_f = factor(expand_f)
print('(f 4 expand) factor: ', factor_f)
```
