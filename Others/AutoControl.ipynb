```
# Based on Mathematical Induction Series
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy.abc import x, i, k, n, I

# Setup Vars
fig = plt.figure(figsize=(18, 12))

# Custom Functions
def calculate(X, y):
  f = lambdify(x, y, 'numpy')
  f = np.vectorize(f)
  return f(X)

# Calculate yprime from
# https://stackoverflow.com/questions/9876290/how-do-i-compute-derivative-using-numpy
def yprime(y):
  return y.diff(x)

def getX0(X, y):
  for index, i in enumerate(X):
    if i == 0:
      return [i, y[index]]
  return None

def Slope0(X, y, yprime):
  y0 = abs(y[0])
  x0 = X[0]
  slope0 = abs(yprime[0])
  for index, slope in enumerate(yprime):
    if abs(slope) < slope0:
      slope0 = abs(slope)
      x0, y0 = X[index], y[index]
  return x0, y0

def outf(f, y):
  print('{:<40} :'.format(str(f)), y)
    
# 1--------------------
# X values
X1 = np.arange(0, 20, 1)

# Math Functions
# sum
# Declare
f_1a = sqrt(pow(x,2) + pow(y, 2)) * exp(I * np.arctan(y/x))
# Calculate
y_1a = calculate(, f_1a)
y_1aprime = calculate(X1, yprime(f_1a))

# exponential
# Declare
f_1b = x**2
# Calculate
y_1b = calculate(X1, f_1b)
y_1bprime = calculate(X1, yprime(f_1b))

# sum of all to square
# Declare
f_1c = (1/2)*(x+(2/x))
# Calculate
y_1c = calculate(X1, f_1c)
y_1cprime = calculate(X1, yprime(f_1c))

# 2--------------------
# X values
X2 = np.arange(0, 10, 1)

# e
# Declare
f_2a = (1+(1/x))**x
# Calculate
y_2a = calculate(X2, f_2a)
y_2aprime = calculate(X2, yprime(f_2a))

# e
# Declare
f_2b = sqrt(x+1)-sqrt(x)
# Calculate
y_2b = calculate(X2, f_2b)
y_2bprime = calculate(X2, yprime(f_2b))

# partil sum O_n
# Declare
f_2c = (1-((0.5)**x))/(1-(0.5))
# Calculate
y_2c = calculate(X2, f_2c)
y_2cprime = calculate(X2, yprime(f_2c))

# factorial
# Declare
f_2d = factorial(x)
# Calculate
y_2d = []
for i in X2:
  y_2d.append(factorial(i))
# y_2dprime = calculate(X2, yprime(f_2d))

# 3--------------------
X3 = np.arange(1, 10, 1)
# Summation
# Declare
f_3a = Sum(x, (x, 1, n)).doit()
# Calculate
y_3a = []
for i in X3:
  y_3a.append(f_3a.subs(n, i))

# Product
# Declare
f_3b = Product(x, (x, 1, n)).doit()
# Calculate
y_3b = []
for i in X3:
  y_3b.append(f_3b.subs(n, i))
  
# Summation with function
# Declare
f_3c = summation(2*x-1, (x, 1, n)).doit()
# Calculate
y_3c = []
for i in X3:
  y_3c.append(f_3c.subs(n, i))
  
# Summation with Complex function
# Declare
f_3d = summation(x**n/factorial(n), (x, 1, n)).doit()
# Calculate
y_3d = []
for i in X3:
  y_3d.append(f_3d.subs(n, i).evalf())


# Results
print("X values:")
print(X1)
print("y values:")
# outf(f_1a, y_1a)
# outf(yprime(f_1a), y_1aprime)
# outf(f_1b, y_1b)
# outf(yprime(f_1b), y_1bprime)
# outf(f_1c, y_1c)
# outf(yprime(f_1c), y_1cprime)
outf(f_3a, y_3a)
outf(f_3b, y_3b)
outf(f_3c, y_3c)
outf(f_3d, y_3d)
# outf(yprime(f_3a), y_3aprime)
```

```
# https://colab.research.google.com/notebooks/charts.ipynb
fig = plt.figure(figsize=(18, 12))

def funcPlot(plt, X, y, p0, clr, fName):
  plt.scatter(X, y, color=clr, s = 10)
  plt.plot(X, y, color=clr)
  plt.scatter(p0[0], p0[1], color='black', marker='^', s = 200)
  plt.title(" function : " + str(fName))
  plt.axhline(linewidth=2, color='black')
  plt.axvline(linewidth=2, color='black')
  plt.xlabel("X")
  plt.ylabel("y")

# Visualising 
plt.subplot2grid((4, 12), (0, 0), rowspan = 1, colspan = 3)
funcPlot(plt, X1, y_1a, getX0(X1, y_1a), 'red', f_1a)
plt.subplot2grid((4, 12), (0, 3), rowspan = 1, colspan = 3)
funcPlot(plt, X1, y_1b, getX0(X1, y_1b), 'blue', f_1b)
plt.subplot2grid((4, 12), (0, 6), rowspan = 1, colspan = 3)  
funcPlot(plt, X1, y_1c, getX0(X1, y_1c), 'green', f_1c)
plt.subplot2grid((4, 12), (0, 9), rowspan = 1, colspan = 3)
funcPlot(plt, X1, y_1a, getX0(X1, y_1a), 'red', '')
funcPlot(plt, X1, y_1b, getX0(X1, y_1b), 'blue', '')
funcPlot(plt, X1, y_1c, getX0(X1, y_1c), 'green', 'All')

plt.subplot2grid((4, 12), (1, 0), rowspan = 1, colspan = 3)
funcPlot(plt, X2, y_2a, Slope0(X2, y_2a, y_2aprime), 'red', f_2a)
plt.subplot2grid((4, 12), (1, 3), rowspan = 1, colspan = 3)
funcPlot(plt, X2, y_2b, Slope0(X2, y_2b, y_2bprime), 'blue', f_2b)
plt.subplot2grid((4, 12), (1, 6), rowspan = 1, colspan = 3)
funcPlot(plt, X2, y_2c, Slope0(X2, y_2c, y_2cprime), 'green', f_2c)
plt.subplot2grid((4, 12), (1, 9), rowspan = 1, colspan = 3)
funcPlot(plt, X2, y_2d, (0,0), 'purple', f_2d)

plt.subplot2grid((4, 12), (2, 0), rowspan = 1, colspan = 3)
funcPlot(plt, X3, y_3a, (0, 0), 'red', f_3a)
plt.subplot2grid((4, 12), (2, 3), rowspan = 1, colspan = 3)
funcPlot(plt, X3, y_3b, (0, 0), 'blue', f_3b)
plt.subplot2grid((4, 12), (2, 6), rowspan = 1, colspan = 3)
funcPlot(plt, X3, y_3c, (0, 0), 'green', f_3c)
plt.subplot2grid((4, 12), (2, 9), rowspan = 1, colspan = 3)
funcPlot(plt, X3, y_3d, (0, 0), 'purple', f_3d)




plt.tight_layout()
plt.show()
```
